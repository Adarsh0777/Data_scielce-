```python
import pandas as pd 
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.DataFrame({"Size":["s","m","l","xxl","xxxl","s","s","m","l","xl","m","xxxl"]})
```

```python
df.head(2)
```

```python
ord_data=[["s","m","l","xl","xxl","xxxl"]]
```

```python
oe=OrdinalEncoder(categories=ord_data)
oe.fit(df[["Size"]])
```

```python
df["Size_en"]=oe.transform(df[["Size"]])
```

```python
df
```

```python
ord={"s":0,"m":1,"l":2,"xl":3,"xxl":4,"xxxl":5}
```

```python
df["Size_enn"]=df["Size"].map(ord)
```

```python
df
```
