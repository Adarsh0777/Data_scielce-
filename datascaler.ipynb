```python
dataset1=pd.read_csv("Electric_Vehicle_Population_Data.csv")
dataset1["Electric Range"].fillna(dataset1["Electric Range"].mean(),inplace=True)
dataset1.isnull().sum()
```

```python
sns.distplot(dataset1["Electric Range"])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python

mms=MinMaxScaler()
mms.fit(dataset1[["Electric Range"]])
```

```python
dataset1["Electric Range_min"]=mms.transform(dataset1[["Electric Range"]])
dataset1
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset1["Electric Range"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset1["Electric Range_min"])
plt.show()


```

```python
dataset2=pd.read_csv("Electric_Vehicle_Population_Data.csv")
dataset2.head(3)
```

```python
dataset2.info()
```

```python
q1=dataset2["Electric Range"].quantile(0.25)
q3=dataset2["Electric Range"].quantile(0.75)
iqr=q3-q1
iqr
```

```python
q1,q3
```

```python
min_r=q1-(1.5*iqr)
max_r=q3+(1.5*iqr)
min_r,max_r
```

```python
dataset2=dataset2[dataset2["Electric Range"]<=max_r]
dataset2
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(dataset2["Electric Range"])
plt.show()
```

```python
from sklearn.preprocessing import FunctionTransformer
import numpy as np 
```

```python
ft=FunctionTransformer(func=np.log1p)
ft.fit([["Electric Range"]])
```

```python
dataset2["Electric Range_ft"]=ft.transform(dataset2["Electric Range"])
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(dataset2["Electric Range"])
plt.title("Before")

plt.subplot(1,2,2)
sns.distplot(dataset2["Electric Range_ft"])
plt.title("After")
plt.show()
```
